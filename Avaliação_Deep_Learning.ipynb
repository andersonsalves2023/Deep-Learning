IMPORTANDO E PROCESSANDO OS DADOS

```
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
```

```
data = pd.read_csv(url,sep=';')
data
```

```
data.describe()
```

```
### lista de números por coluna
for i in data.keys():
  print(f'{i} has {len(np.unique(data[i]))} unique values')
```

```
# gráficos de pares
col2plot = ['fixed acidity', 'volatile acidity','citric acid','quality']
sns.pairplot(data[col2plot],kind='reg',hue='quality')
plt.show()
```

```
## visualizando outliers
```

```
fig,ax = plt.subplots(1,figsize=(18,5))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()


```

```
## removendo os outliers
data = data[data['total sulfur dioxide']<200]

fig,ax = plt.subplots(1,figsize=(18,5))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()
```

```
## normalização dos dados (exceto qualidade)
cols2zscore = data.keys()
cols2zscore = cols2zscore.drop('quality')

## escrito para maior clareza
for col in cols2zscore:
  meanval   = np.mean(data[col])
  stdev     = np.std(data[col],ddof=1)
  data[col] = (data[col]-meanval) / stdev

data.describe()
```

```
## verificando a plotagem novamente
fig,ax = plt.subplots(1,figsize=(18,5))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()

## gráficos no mesmo range
```

```
## Distribuição das qualidades dos dados
fig = plt.figure(figsize=(9,6))
plt.rcParams.update({'font.size': 22})

counts = data['quality'].value_counts()
plt.bar(list(counts.keys()),counts)
plt.xlabel('Quality rating')
plt.ylabel('Count')
plt.show

# criando uma nova coluna para qualidade binária
data['boolQuality'] = 0
data['boolQuality'][data['quality']>5] = 1

data[['quality','boolQuality']]
```
